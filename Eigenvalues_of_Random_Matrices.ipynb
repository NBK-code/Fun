# Eigenvalues of Random Matrices

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dim = 3
M = np.random.rand(dim, dim)
M
```

```python
dim = 3
M = (np.random.rand(dim, dim) > 0.5)
M
```

```python
dim = 3
M = (np.random.rand(dim, dim) > 0.5).astype(float)
M
```

```python
dim = 3
M = (np.random.rand(dim, dim) > 0.5).astype(float) - 0.5
M
```

```python
dim = 3
M = ((np.random.rand(dim, dim) > 0.5).astype(float) - 0.5)*2
M
```

```python
eigenvalues, _ = np.linalg.eig(M)
eigenvalues
```

```python
np.real(eigenvalues), np.imag(eigenvalues)
```

##Dim = 2

```python
dim = 2
S = 5000

for _ in range(S):
  M = M = ((np.random.rand(dim, dim) > 0.5).astype(float) - 0.5)*2
  eigenvalues, _ = np.linalg.eig(M)
  plt.scatter(np.real(eigenvalues), np.imag(eigenvalues), color = 'red', marker = '+', alpha = 0.5)
```
